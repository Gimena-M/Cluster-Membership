# Support Vector Machine Classification

```python
# i don't use cudf yet, since fitting for gridsearchcv takes pandas df and not cudf df............
# this way, i can still use imbalanced-learn
```

```python
# read galaxies table
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Cluster-Membership-DATA/HSC-unWISE-W03.csv')

# with cudf
# from cudf import read_csv
# df = read_csv('/content/drive/MyDrive/Cluster-Membership-DATA/HSC-unWISE-W01.csv')
```

```python
# features and labels
with open('/content/drive/MyDrive/Cluster-Membership-DATA/all_features.txt') as file:
    features = file.read().splitlines()
labels = 'member'
```

```python
# Train/Test Split
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size = 0.3, stratify = df[labels], random_state= 42)

# with cuml
# from cuml.model_selection import train_test_split
# train_feat, test_feat, train_lab, test_lab = train_test_split(X= df[features], y=df[labels],
#                                                             stratify= df[labels], test_size = 0.3, random_state = 42)
```

```python
# Undersample majority class
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy= 5./5., replacement= False)
rus_feat, rus_lab = rus.fit_resample(train[features], train[labels])
rus_feat[labels] = rus_lab
train = rus_feat
del rus, rus_feat, rus_lab

n_mem = train[train.member == 1].shape[0]
n_no = train[train.member == 0].shape[0]
n = train.shape[0]
print ('Members: {} ({:.2f}%)'.format(n_mem, n_mem/n*100))
print ('Non members: {} ({:.2f}%)'.format(n_no, n_no/n*100))
del n,n_mem,n_no
```

```python
# SVM model
from cuml.svm import SVC
svc = SVC(class_weight= 'balanced', cache_size = 4000, verbose = False)
```

```python
# random search
params = [
    {"kernel": ["linear"], "C": [0.1, 1, 10]},
    {"kernel": ["poly"], "C": [0.1, 1, 10], 'degree' : [2,3,4,7], "gamma": ['scale', 0.005, 0.1, 1.]},
    {"kernel": ["rbf"], "C": [0.1, 1, 10], "gamma": ['scale', 0.005, 0.1, 1.]},
  ]

from cuml.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = svc, param_grid= params, cv= 3, n_jobs= -1,  verbose= 4, scoring = 'f1')
grid_search.fit(train[features], train[labels])
```

```python
svc = grid_search.best_estimator_
svc.get_params()
```

```python
import cudf
pred = svc.predict(cudf.from_pandas(test[features]))
```

```python
from sklearn.metrics import classification_report
print(classification_report(test[labels], pred.to_numpy()))
```

```python
df_gs = pd.DataFrame(grid_search.cv_results_).sort_values('rank_test_score')
df_gs.drop(columns= 'params').tail(30)
```
