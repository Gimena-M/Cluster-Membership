```python
"""
Make subsets with limits in some feature and check z and mass distributions
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../clean-HSC-unWISE-W03.csv')
```

```python
feat = 'i_cmodel_mag'
val = 21

cond = df[feat] <= val
plt.figure(figsize=(15,4))

plt.subplot(1, 3, 1)
sns.kdeplot(data = df[df.member == 0], x = feat, fill= True, alpha = 0.5, color = 'teal', label = 'Non member')
sns.kdeplot(data = df[df.member == 1], x = feat, fill= True, alpha = 0.5, color = 'crimson', label = 'Member')
plt.grid()
plt.legend()

plt.subplot(1, 3, 2)
sns.kdeplot(data = df, x = 'phot_z', fill= True, alpha = 0.5, color = 'violet', label = 'No limits')
sns.kdeplot(data = df[cond], x = 'phot_z', fill= True, alpha = 0.5, color = 'chocolate', label = f'{feat} < {val}')
plt.legend()
plt.grid()

plt.subplot(1, 3, 3)
sns.kdeplot(data = df, x = 'log_st_mass', fill= True, alpha = 0.5, color = 'violet', label = 'No limits')
sns.kdeplot(data = df[cond], x = 'log_st_mass', fill= True, alpha = 0.5, color = 'chocolate', label = f'{feat} < {val}')
plt.legend()
plt.grid()
```

```python
feat = 'i_sdssshape_shape11sigma'
val = 0.005

cond = df[feat] <= val
plt.figure(figsize=(15,4))

plt.subplot(1, 3, 1)
sns.kdeplot(data = df[df.member == 0], x = feat, fill= True, alpha = 0.5, color = 'teal', label = 'Non member')
sns.kdeplot(data = df[df.member == 1], x = feat, fill= True, alpha = 0.5, color = 'crimson', label = 'Member')
plt.grid()
plt.legend()
plt.xlim(0, 0.05)

plt.subplot(1, 3, 2)
sns.kdeplot(data = df, x = 'phot_z', fill= True, alpha = 0.5, color = 'violet', label = 'No limits')
sns.kdeplot(data = df[cond], x = 'phot_z', fill= True, alpha = 0.5, color = 'chocolate', label = f'{feat} < {val}')
plt.legend()
plt.grid()

plt.subplot(1, 3, 3)
sns.kdeplot(data = df, x = 'log_st_mass', fill= True, alpha = 0.5, color = 'violet', label = 'No limits')
sns.kdeplot(data = df[cond], x = 'log_st_mass', fill= True, alpha = 0.5, color = 'chocolate', label = f'{feat} < {val}')
plt.legend()
plt.grid()
```
