```python
"""
Make subsets with limits in some feature and check z and mass distributions
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../HSC-unWISE-W03.csv')
```

```python
def func(cond, feat, label):
    plt.figure(figsize=(15,4))

    plt.subplot(1, 3, 1)
    sns.kdeplot(data = df[df.member == 0], x = feat, fill= True, alpha = 0.5, color = 'teal', label = 'Non member')
    sns.kdeplot(data = df[df.member == 1], x = feat, fill= True, alpha = 0.5, color = 'crimson', label = 'Member')
    plt.grid()
    plt.legend()

    plt.subplot(1, 3, 2)
    sns.kdeplot(data = df, x = 'phot_z', fill= True, alpha = 0.5, color = 'violet', label = 'No limits')
    sns.kdeplot(data = df[cond], x = 'phot_z', fill= True, alpha = 0.5, color = 'chocolate', label = label)
    plt.legend()
    plt.grid()

    plt.subplot(1, 3, 3)
    sns.kdeplot(data = df, x = 'log_st_mass', fill= True, alpha = 0.5, color = 'violet', label = 'No limits')
    sns.kdeplot(data = df[cond], x = 'log_st_mass', fill= True, alpha = 0.5, color = 'chocolate', label = label)
    plt.legend()
    plt.grid()
```

```python
val = 21
feat = 'i_cmodel_mag'
func(cond = df[feat] <= val , feat = 'i_cmodel_mag', label = f'{feat} < {val}')
```

```python
feat = 'i_sdssshape_shape11sigma'
val = 0.005
func(cond = df[feat] <= val , feat = feat, label = f'{feat} < {val}')
```

```python
feat = 'gr'
min = -99
max = 1
func(cond = (df[feat] >= min) & (df[feat] <= max) , feat = feat, label = f'{feat} < {val}')
```

```python
feat = 'ri'
min = -99
max = 1
func(cond = (df[feat] >= min) & (df[feat] <= max) , feat = feat, label = f'{feat} < {val}')
```
