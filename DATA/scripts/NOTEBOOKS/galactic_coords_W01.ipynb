```python
import pandas as pd
import matplotlib.pyplot as plt

import astropy.units as u
from astropy.coordinates import SkyCoord
```

```python
df = pd.read_table('../Wen+Han/HSC-unWISE-W01.dat', sep='\s+', usecols=[0,1,12,13,14,15,16,17,18], names=['ra','dec', 'W1', 'W1_err', 'W2', 'W2_err', 'phot_z', 'phot_z_err', 'log_st_mass'])
```

```python
c = SkyCoord(df.ra.values*u.degree, df.dec.values*u.degree)
```

```python
gc = c.transform_to('galactic')
```

```python
plt.scatter(gc.l.value, gc.b.value, alpha = 0.1, marker = 'x', s = 2)
```
